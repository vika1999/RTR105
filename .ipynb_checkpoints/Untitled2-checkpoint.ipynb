```python
#Laboratorijas darbs 170
(Aptuveni novērtējot funkcijas saknes vērtību lietojot ”peli” iegūstam x=3,1346.)
from math import sin, fabs
from time import sleep

def f(x):
    return sin(x)
k=0
a = 1.1
b = 3.2

funa = f(a)
funb = f(b)

if (funa*funb>0.0):
    print ("Funkcijas sin(x) dotaja intervala [%s, %s] saknju nav" %(a,b))
    sleep(1); exit()
else:
    print ("Funkcijas sin(x) dotaja intervala sakne(s) ir!")

deltax = 0.001

while (fabs(b-a)>deltax):
    k=k+1
    x=(a+b)/2; funx=f(x)
    if (funa*funx < 0.):
        b=x       
    else:
        a=x
print ("sin(x) Sakne ir:", x)
print ("y=", sin(x))
print ("k=", k)

Funkcijas sin(x) dotaja intervala sakne(s) ir!
sin(x) Sakne ir: 3.1420654296875004
y= -0.00047277608009484905
k= 12
```

```python
#Aptuveni novērtējot funkcijas saknes vērtību lietojot ”peli” iegūstam x=0.0109052

from math import atan, fabs
from time import sleep

def f(x):
    return atan(x)
k=0
a = -3
b = 3

funa = f(a)
funb = f(b)

if (funa*funb>0.0):
    print ("Funkcijas atan(x) dotaja intervala [%s, %s] saknju nav" %(a,b))
    sleep(1); exit()
else:
    print ("Funkcijas atan(x) dotaja intervala sakne(s) ir!")

deltax = 0.0001

while (fabs(b-a)>deltax):
    k=k+1
    x=(a+b)/2; funx=f(x)
    if (funa*funx < 0.):
        b=x       
    else:
        a=x
print ("atan(x) sakne ir:", x)
print ("f(x)=", atan(x))
print ("k=", k)


Funkcijas atan(x) dotaja intervala sakne(s) ir!
atan(x) sakne ir: 9.1552734375e-05
f(x)= 9.155273411920462e-05
k= 16
```
